```python
from cryptography.hazmat.primitives.ciphers import Cipher, algorithms, modes
from cryptography.hazmat.backends import default_backend
from cryptography.hazmat.primitives import padding
import base64
import os
from cryptography.hazmat.primitives import serialization, padding, hashes
from cryptography.hazmat.primitives.asymmetric import padding as asymmetric_padding
from cryptography.hazmat.primitives.asymmetric import rsa
from cryptography.hazmat.backends import default_backend
import base64
import hmac
import hashlib
from Crypto.PublicKey import RSA
from Crypto.Signature import pkcs1_15
from Crypto.Hash import SHA256
from base64 import b64encode, b64decode



# Digital Signature Generation using RSA
def generate_signature(message, private_key):
    hash_obj = SHA256.new(message.encode())
    signature = pkcs1_15.new(private_key).sign(hash_obj)
    return b64encode(signature).decode('utf-8')

# Digital Signature Verification using RSA
def verify_signature(message, signature, public_key):
    hash_obj = SHA256.new(message.encode())
    try:
        pkcs1_15.new(public_key).verify(hash_obj, b64decode(signature))
        return True
    except (ValueError, TypeError):
        return False


def aes_encrypt_cbc(key, iv, plaintext):
    cipher = Cipher(algorithms.AES(key), modes.CBC(iv), backend=default_backend())
    encryptor = cipher.encryptor()

    # Apply padding to plaintext
    padder = padding.PKCS7(128).padder()
    padded_data = padder.update(plaintext) + padder.finalize()

    ciphertext = encryptor.update(padded_data) + encryptor.finalize()
    return ciphertext

def aes_decrypt_cbc(key, iv, ciphertext):
    cipher = Cipher(algorithms.AES(key), modes.CBC(iv), backend=default_backend())
    decryptor = cipher.decryptor()
    decrypted_padded_data = decryptor.update(ciphertext) + decryptor.finalize()

    # Remove padding from decrypted data
    unpadder = padding.PKCS7(128).unpadder()
    plaintext = unpadder.update(decrypted_padded_data) + unpadder.finalize()

    return plaintext

def AES():
    while True:
        print("AES CBC Encryption/Decryption")
        print("1. Encrypt and Decrypt")
        print("2. Exit")

        choice = input("Enter your choice: ")

        if choice == "1":
            mode = "CBC"
            operation = input("Enter operation (Encrypt/Decrypt): ").lower()
            data = input("Enter plaintext/ciphertext: ").encode()
            key = os.urandom(16)  # Generate random 16-byte key
            iv = os.urandom(16)   # Generate random 16-byte IV

            if operation == "encrypt":
                ciphertext = aes_encrypt_cbc(key, iv, data)
                decrypted_text = aes_decrypt_cbc(key, iv, ciphertext)
                print("Key:", key.hex())
                print("IV:", iv.hex())
                print("Encrypted text:", base64.b64encode(ciphertext).decode())
                print("Decrypted text:", decrypted_text.decode())
            elif operation == "decrypt":
                plaintext = aes_decrypt_cbc(key, iv, data)
                encrypted_text = aes_encrypt_cbc(key, iv, plaintext)
                print("Key:", key.hex())
                print("IV:", iv.hex())
                print("Decrypted text:", plaintext.decode())
                print("Encrypted text:", base64.b64encode(encrypted_text).decode())
            else:
                print("Invalid operation.")
        elif choice == "2":
            print("Exiting...")
            break
        else:
            print("Invalid choice. Please choose a valid option.")



def rsa_encrypt(public_key, plaintext):
    ciphertext = public_key.encrypt(
        plaintext,
        asymmetric_padding.OAEP(
            mgf=asymmetric_padding.MGF1(algorithm=hashes.SHA256()),
            algorithm=hashes.SHA256(),
            label=None
        )
    )
    return ciphertext

def rsa_decrypt(private_key, ciphertext):
    plaintext = private_key.decrypt(
        ciphertext,
        asymmetric_padding.OAEP(
            mgf=asymmetric_padding.MGF1(algorithm=hashes.SHA256()),
            algorithm=hashes.SHA256(),
            label=None
        )
    )
    return plaintext

def generate_rsa_key_pair():
    private_key = rsa.generate_private_key(public_exponent=65537, key_size=2048, backend=default_backend())
    return private_key, private_key.public_key()

def RSA():
    print("RSA Encryption/Decryption")

    private_key, public_key = generate_rsa_key_pair()

    while True:
        operation = input("Enter operation (Encrypt/Decrypt) or 'exit' to quit: ").lower()

        if operation == "exit":
            print("Exiting...")
            break

        if operation != "encrypt" and operation != "decrypt":
            print("Invalid operation. Please choose 'Encrypt' or 'Decrypt'.")
            continue

        data = input("Enter plaintext/ciphertext: ").encode()

        if operation == "encrypt":
            ciphertext = rsa_encrypt(public_key, data)
            print("Encrypted text:", base64.b64encode(ciphertext).decode())
        else:
            ciphertext = base64.b64decode(data)
            decrypted_text = rsa_decrypt(private_key, ciphertext)
            print("Decrypted text:", decrypted_text.decode())


import hashlib

def hash_data(data, algorithm):
    if algorithm == "sha1":
        hasher = hashlib.sha1()
    elif algorithm == "sha256":
        hasher = hashlib.sha256()
    else:
        raise ValueError("Invalid hash algorithm")

    hasher.update(data)
    hash_value = hasher.hexdigest()
    return hash_value

def HASH():
    print("Hashing")

    while True:
        data = input("Enter data to be hashed (or 'exit' to quit): ")
        if data.lower() == "exit":
            print("Exiting...")
            break

        hash_mode = input("Enter hash mode (SHA1 / SHA256): ").lower()

        try:
            hash_result = hash_data(data.encode(), hash_mode)
            print(f"{hash_mode.upper()} Hash Value:", hash_result)
        except ValueError as e:
            print(e)




def generate_mac(message, key):
    mac = hmac.new(key.encode(), message.encode(), hashlib.sha256)
    return mac.hexdigest()



from Crypto.PublicKey import RSA
from Crypto.Signature import pkcs1_15
from Crypto.Hash import SHA256
from base64 import b64encode, b64decode

# Digital Signature Generation using RSA
def generate_signature(message, private_key):
    hash_obj = SHA256.new(message.encode())
    signature = pkcs1_15.new(private_key).sign(hash_obj)
    return b64encode(signature).decode('utf-8')

# Digital Signature Verification using RSA
def verify_signature(message, signature, public_key):
    hash_obj = SHA256.new(message.encode())
    try:
        pkcs1_15.new(public_key).verify(hash_obj, b64decode(signature))
        return True
    except (ValueError, TypeError):
        return False

def s_rsa():
    rsa_key = RSA.generate(2048)
    private_key = rsa_key.export_key()
    public_key = rsa_key.publickey().export_key()

    private_key_obj = RSA.import_key(private_key)
    public_key_obj = RSA.import_key(public_key)

    print("Digital Signature using RSA:")

    message = input("Enter the message: ")

    # Generate signature
    signature = generate_signature(message, private_key_obj)
    print("Generated Signature:", signature)

    # Verify signature
    is_verified = verify_signature(message, signature, public_key_obj)
    if is_verified:
        print("Signature is valid.")
    else:
        print("Signature is not valid.")




print("Which cryptographic function you want to try?")

print("AES Encryption/Decryption - press 1")
print("RSA Encryption/Decryption - press 2")
print("Hashing - press 3")
print("Digital Signature using RSA - press 4")
print("MAC Generation: - press 5")

inpt = int(input())

if(inpt==1) :
    AES()
elif(inpt==2) :
    RSA()
elif(inpt==3) :
    HASH()
elif(inpt==4) :
    s_rsa()
elif(inpt==5) :

    secret_key = "abcd"
    message = input("Enter a plaintext: ")

    mac = generate_mac(message, secret_key)

    print("Message Authentication Code (HMAC-SHA256):")
    print("Message:", message)
    print("Generated MAC:", mac)
```
